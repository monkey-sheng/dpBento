```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

sns.set_theme('paper', font_scale=1.35, style='whitegrid', rc={"axes.titlesize": 18, "axes.labelsize": 16,})
```

```python
# compute micro benchmark - int8

x_name = 'Operation Type'
# x_val = ['add', 'addi', 'sub', 'mul', 'div', ]
x_val = ['add', 'sub', 'mul', 'div']
y_name = 'Billion Ops/s'

var_name = 'DPU'

# int32 results
bf3_vals_int8 = np.array([1704759620,1705588140,1389225440,494793440]) / 1E9
host_vals_int8 = np.array([6535709460,6429568816,2760961472,821801996]) / 1E9
bf2_vals_int8 = np.array([1442316296,1376416920,1245797800,792887232]) / 1E9

# BF-2 using dub data
raw_data = pd.DataFrame.from_dict({x_name: x_val, 'BF-2': bf2_vals_int8, 'BF-3': bf3_vals_int8, 'Host CPU': host_vals_int8})
data = pd.melt(raw_data, id_vars=x_name, value_name=y_name, var_name=var_name)

plt.figure(figsize=(8, 5))
ax = sns.barplot(data=data, x=x_name, y=y_name, hue=var_name, gap=0.12)
# display the value on top of the bar
for p in ax.patches:
    if p.get_height() > 0:
        ax.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 10), textcoords='offset points', size=14, weight='bold')
# ax.set_yscale("log")
ax.set_xticklabels(ax.get_xticklabels(), size=18)
ax.set_yticklabels(ax.get_yticklabels(), size=18, weight='bold')
ax.set_title('int8 Primitive Operations Throughput', size=20, weight='bold', pad=20)

plt.legend(fontsize=13.8, loc='upper right',)
plt.tight_layout()
# plt.show()

plt.savefig('compute-int8.pdf')
```
